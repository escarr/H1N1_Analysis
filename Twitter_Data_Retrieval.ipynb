### Note: tweets were retrieved manually--I changed the date range for each API call as well as which tweet_dict_timeX I was writing the results to and the output file

```python
# define query
# could be easily changed to support other similar projects by changing H1N1 and swine flue to another infection

PRODUCT = 'fullarchive'
LABEL = 'dev1'
r = api.request('tweets/search/%s/:%s' % (PRODUCT, LABEL), 
        {'query':'lang:en (H1N1 OR swine flu) (vaccine OR vaccines OR vaccination OR vaccinations OR vaccinate OR vaccinates OR vaccinated) -RT', 
         'fromDate':"201002220000", 'toDate':"201002230000"})
```

```python
# write tweets to dictionary: include 'id', 'created_at' and 'text'
import json    
count = 0
for item in r:
    count += 1
    tweet_id = item['id'] # unique identifier 
    time = item['created_at'] # time stamp
    text = item['text'] # tweet text
    tweet_dict_time8[tweet_id] = {"timestamp":time, "text":text}
    
print(count)
```
